```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/Housing_Boston.csv")
df.head()
```

- **RM** = average number of rooms per dwellin
- **LSTAT** = % lower status of the population
- **MEDV** = Median value of owner-occupied homes in $1000's

**¿Existe un modelo matemático que pueda predecir el valor de la vivienda en Boston?**

```python
y = df['medv']
X = df[['rm']]
```

**¿Cómo sabemos el máximo y el mínimo?**

```python
print(max(X['rm']), min(X['rm']))
print(max(y), min(y))
```

**¿Cómo sabemos la cantidad de datos?**

```python
print(len(X))
```

**Entrenar el modelo y mostrar datos**
- A continuación, queremos ajustar un modelo de regresiión líneal, utilizando el modelo de mínimos cuadrados ordinarios

```python
model = sm.OLS(y,X).fit()
```

Con esta instrucción, entrenamos el modelo con los datos de la variable independiente X, así como, los correspondientes valores de la variable dependiente y

```python
print(model.summary())
```

El coeficiente de 3.6534 significa que a medida que la variable RM auemnta en 1, el valor a predecir de MDEV aumenta en 3.6534

El RM tiene un valor de P estadísticamente significativo

Hay un intervalo de confianza del 95% para el RM (lo que significa que predecimos con un 95% de confianza que el valor de RM está entre 3.548 y 3.759)

Es importantte tener en cuenta que Statsmodels no agrega una constante de forma predeterminada

Hay suficiente evidencia estadística para rechazar la hipótesis nula

```python
X = sm.add_constant(X)  # Agregar una constante (intercepto) al modelo
```

```python
model = sm.OLS(y,X).fit()
print(model.summary())
```
