```python
# Necesito saber la proporción de robos con violencia y sin violencia por tipo de robo y por día de la semana
proporcion_violencia = df.groupby(['TIPO', 'DIA_DE_LA_SEMANA', 'VOLENCIA']).size().unstack(fill_value=0)
proporcion_violencia['TOTAL'] = proporcion_violencia.sum(axis=1)
proporcion_violencia['PROPORCION_SI'] = proporcion_violencia['SI'] / proporcion_violencia['TOTAL']
proporcion_violencia['PROPORCION_NO'] = proporcion_violencia['NO'] / proporcion_violencia['TOTAL']
proporcion_violencia
```

```python
# Ayúdame a graficar esta tabla pero solo con la violencia "SI" en un heatmap
plt.figure(figsize=(12,6))
sns.heatmap(proporcion_violencia.pivot_table(index='TIPO', columns='DIA_DE_LA_SEMANA', values='PROPORCION_SI'), annot=True, fmt=".2f", cmap="YlGnBu")
plt.title('Proporción de robos con violencia (SI) por tipo y día de la semana')
plt.show()
```

```python
# Ahora por ROBO A NEGOCIO hazlo en proporción por la hora del día con ciolencia SI
df_robo_negocio = df[df['TIPO'] == 'ROBO A NEGOCIO']
proporcion_hora_violencia = df_robo_negocio.groupby(['HORA', 'VOLENCIA']).size().unstack(fill_value=0)
proporcion_hora_violencia['TOTAL'] = proporcion_hora_violencia.sum(axis=1)
proporcion_hora_violencia['PROPORCION_SI'] = proporcion_hora_violencia['SI'] / proporcion_hora_violencia['TOTAL']
proporcion_hora_violencia['PROPORCION_NO'] = proporcion_hora_violencia['NO'] / proporcion_hora_violencia['TOTAL']
proporcion_hora_violencia
```

```python
# Crear un indice de riesgo de 0 a 100 que involucre frecuencia de los eventos, violencia y hora del día
# Definir 3 franjas horarias: mañana (6-14), tarde (14-22), noche (22-6)
def asignar_franja(hora):
    if 6 <= hora < 14:
        return 'mañana'
    elif 14 <= hora < 22:
        return 'tarde'
    else:
        return 'noche'

# Crear columna de franja horaria
df['FRANJA'] = df['HORA'].apply(asignar_franja)

# Calcular frecuencia de eventos por franja y tipo de violencia
eventos_franja = df.groupby(['FRANJA', 'VOLENCIA']).size().reset_index(name='FRECUENCIA')

# Normalizar frecuencia a 0-100
max_frec = eventos_franja['FRECUENCIA'].max()
eventos_franja['FRECUENCIA_NORM'] = (eventos_franja['FRECUENCIA'] / max_frec) * 100

# Asignar puntaje de violencia: SI=100, NO=0
eventos_franja['VIOLENCIA_PUNT'] = eventos_franja['VOLENCIA'].map({'SI': 100, 'NO': 0})

# Índice de riesgo: promedio ponderado (70% violencia, 30% frecuencia)
eventos_franja['INDICE_RIESGO'] = 0.7 * eventos_franja['VIOLENCIA_PUNT'] + 0.3 * eventos_franja['FRECUENCIA_NORM']

# Mostrar la tabla de índice de riesgo por franja y violencia
eventos_franja[['FRANJA', 'VOLENCIA', 'INDICE_RIESGO']]
```

```python
# Usamos folium para graficar los puntos en un mapa del dataframe df_filtro
import folium
from folium.plugins import MarkerCluster  
## esta ubicación debe ser cambiada por la ubicación de Chihuahua, chih. 
mapa = folium.Map(location=[28.6325, -106.0691], zoom_start=12)
marker_cluster = MarkerCluster().add_to(mapa)
for idx, row in df_filtro.iterrows():
    folium.Marker(location=[row['LATITUD'], row['LONGITUD']],
                  popup=f"FOLIO: {row['FOLIO']}<br>FECHA: {row['FECHA']}<br>HORA: {row['HORA']}<br>TIPO: {row['TIPO']}<br>DISTRITO: {row['DISTRITO']}<br>CUADRANTE: {row['CUADRANTE']}").add_to(marker_cluster)
mapa
```

```python
# Con df_filtro genera una mapa de calor con folium
from folium.plugins import HeatMap
mapa_calor = folium.Map(location=[28.6325, -106.0691], zoom_start=12)
heat_data = [[row['LATITUD'], row['LONGITUD']] for index, row in df_filtro.iterrows()]
HeatMap(heat_data).add_to(mapa_calor)
mapa_calor
```
