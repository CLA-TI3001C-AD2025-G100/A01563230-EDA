```python
def add_violence_cols(cuadrante_df, source_df):
    # normalizar texto y agrupar por CUADRANTE contando SI y NO
    s = source_df['VIOLENCIA'].fillna('').astype(str).str.strip().str.upper()
    agg = source_df.assign(VIOL=s).groupby('CUADRANTE').agg(
        CON_VIOLENCIA = ('VIOL', lambda x: (x == 'SI').sum()),
        SIN_VIOLENCIA = ('VIOL', lambda x: (x == 'NO').sum())
    ).reset_index()
    # merge con el dataframe de cuadrantes y completar con 0
    res = pd.merge(cuadrante_df, agg, on='CUADRANTE', how='left')
    res['CON_VIOLENCIA'].fillna(0, inplace=True)
    res['SIN_VIOLENCIA'].fillna(0, inplace=True)
    res['CON_VIOLENCIA'] = res['CON_VIOLENCIA'].astype(int)
    res['SIN_VIOLENCIA'] = res['SIN_VIOLENCIA'].astype(int)
    return res

cuadrante_negocio = add_violence_cols(cuadrante_negocio, df_negocio)
cuadrante_casa = add_violence_cols(cuadrante_casa, df_casa)
cuadrante_vehiculo = add_violence_cols(cuadrante_vehiculo, df_calle)

```
